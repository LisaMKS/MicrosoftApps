```python
import win32com.client
import pandas as pd

def scrape_attendees_by_partial_name(partial_name):
    # Create an instance of the Outlook application
    outlook = win32com.client.Dispatch("Outlook.Application")

    # Get the default Outlook session
    namespace = outlook.GetNamespace("MAPI")

    # Get the calendar folder
    calendar = namespace.GetDefaultFolder(9)  # 9 represents the Calendar folder

    # Find the desired meeting by partial name
    meetings = calendar.Items
    meetings.Sort("[Start]")
    meetings.IncludeRecurrences = True

    attendees_data = []

    for meeting in meetings:
        if partial_name.lower() in meeting.Subject.lower():
            # Get the list of attendees
            attendees = meeting.Recipients

            for attendee in attendees:
                # Resolve the recipient's name and get the corresponding contact
                attendee.Resolve()
                contact = namespace.CreateRecipient(attendee.Name)
                contact = contact.AddressEntry.GetExchangeUser()

                # Get the job title and line manager
                job_title = contact.JobTitle
                line_manager = contact.GetExchangeUserManager().Name

                # Add attendee's information to the list
                attendee_info = {
                    "Name": contact.Name,
                    "Job Title": job_title,
                    "Line Manager": line_manager
                }
                attendees_data.append(attendee_info)

    if len(attendees_data) == 0:
        print("No meeting found with the partial name: '{}'".format(partial_name))
        return

    # Create a DataFrame from the attendees' data
    df = pd.DataFrame(attendees_data)

    # Export the DataFrame to an Excel file
    file_name = "CEG Reps and Exec List.xlsx"
    df.to_excel(file_name, index=False)
    print("Exported attendee information to '{}'".format(file_name))

# Call the function with the partial meeting name
scrape_attendees_by_partial_name("Connect and Evolve")
```
